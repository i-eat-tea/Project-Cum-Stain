```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import bs4
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Provinces_of_Cambodia" 
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=header)
soup = bs4.BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
rows = table.find_all('tr')[1:]
data = []
for row in rows:
    cols = row.find_all('td')
    if len(cols) > 1:
        name = cols[1].text.strip().replace(',', '')
        capital = cols[4].text.strip().replace(',', '')
        population = cols[5].text.strip().replace(',', '')
        area = cols[6].text.strip().replace(',', '')
        density = cols[7].text.strip().replace(',', '')
        iso_code = cols[8].text.strip().replace(',', '')

        data.append({'Province': name, 'Capital': capital, 'Population': population, 'Area': area, 'Density': density, 'ISO Code': iso_code})

df = pd.DataFrame(data)
```
